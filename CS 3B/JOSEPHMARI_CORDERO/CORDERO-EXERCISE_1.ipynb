(20 points) Determine whether the student passes (1) or fails (0) based on:
Hours studied
Hours of sleep
Hours of free time
Given weights w1 = 0.6, w2, 0.4, bias = -3
Step function with a threshold of 1 

Predict the output of the following inputs:
(x1, x2) = (8, 7)
(x1, x2) = (3, 4)

```python
weights = [0.6, 0.4]
bias = -3

# For the first set of inputs

inputs1 = [8, 7]
output1 = 0

for i in range(len(inputs1)):
    output1 += inputs1[i] * weights[i]

print("Output: ", output1)

output1 += bias
print("Output with bias: ", output1)

activate = output1 > 1
print("Activation: ", activate)

print("\n==============================================\n")

# For the second set of inputs

inputs2 = [3, 4]
output2 = 0

for i in range(len(inputs2)):
    output2 += inputs2[i] * weights[i]

print("Output: ", output2)

bias = -3
output2 += bias
print("Output with bias: ", output2)

activate = output2 > 1
print("Activation: ", activate)
```

(20 points) Logic Gate Simulation. Given the following setup for a perceptron, compute its output and verify whether it acts as an AND gate.
Given weights w1 = 1, w2 = 1, bias = -1.5
Step function with a threshold of 0
Inputs	:
	(0,0) , (0,1), (1,0), (1,1)


```python
weights = [1, 1]
bias = -1.5

inputs = [ [0,0], [0,1], [1,0], [1,1] ]

for input in inputs:
    output = 0
    for i in range(len(weights)):
        product = input[i] * weights[i]
        output += product
    print("Output:", output)
    output += bias
    print("Output with bias:", output)
    activate = output > 0
    print("Activation:", activate)
    print("\n==============================================\n")
```

^^^^^^^^^ Based on the reults, it does act as an AND gate. This Logic Gate Simulation is successful.

(60 points) Perceptron comparison (One vs All). Given the 3 perceptron, using the same input, compute the output and decided on the predicted class is the WINNER. If a tie is present, compare and get the highest weighted sum. 


Inputs = [0.5, -1, 2, 1, 0]
Perceptron A Configuration:
Weights: WA = [1.0, -0.5, 0.2, 0.1, 0.0]
BiasA: 0.2
Perceptron B
Weights: WB = [0.2, 0.2, 0.5, -0.4, 0.3]
BiasB: 0.0
Perceptron C
Weights: WC = [-0.3, -0.1, 0.4, 0.0, 0.2]
BiasC: -0.6

```python
inputs = [0.5, -1, 2, 1, 0]
weights = [
    [1.0, -0.5, 0.2, 0.1, 0.0],    
    [0.2, 0.2, 0.5, -0.4, 0.3],    
    [-0.3, -0.1, 0.4, 0.0, 0.2]   
]
bias = [0.2, 0.0, -0.6]

outputs = []
for p in range(3):
    output = sum(inputs[i] * weights[p][i] for i in range(len(inputs)))
    output += bias[p]
    outputs.append(output)
    print(f"Perceptron {chr(65+p)} Output: {output}")
    

# Find the winner
active = [i for i, o in enumerate(outputs) if o > 0]
if len(active) == 1:
    print(f"\nWinner: Perceptron {chr(65+active[0])}")
elif len(active) > 1:
    winner = active[max(range(len(active)), key=lambda x: outputs[active[x]])]
    print(f"\nTie! Winner by highest output: Perceptron {chr(65+winner)}")
else:
    print("\nNo perceptron activated (output <= 0 for all).")
```
